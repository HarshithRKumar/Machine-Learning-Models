```python
import pandas as pd
df=pd.read_csv('/content/abalone cluster data.csv')
df.head()
```

```python
df=df.drop(columns=['Type','LongestShell','WholeWeight','ShuckedWeight','VisceraWeight','ShellWeight','Rings'])
df.head()
```

**Agglomerative**

```python
df=df.iloc[:100]
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,linkage='ward')# ward method
cluster.fit_predict(df)
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(df,cluster.labels_)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Diameter'],df['Height'],c=cluster.labels_)
plt.show()
```

**Divisive**



```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

df=df.iloc[:10]
Z = linkage(df, 'single') # single Distance

dendrogram(Z) #plotting the dendogram

plt.title('Dendrogram')
plt.show()
```
