```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

```

```python
df=pd.read_csv("/content/data.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
column=['date','street','city','statezip','country']
for col in column:
  df[col]=encoder.fit_transform(df[col])
df.head(10)
```

```python
df.isnull().sum()
```

```python
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
x=df.drop("price",axis=1)
y=df["price"]

scaler=StandardScaler()
x=scaler.fit_transform(x)

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

model=LinearRegression()
model.fit(x_train,y_train)

pred=model.predict(x_test)

r2=r2_score(y_test,pred)
print("lr r2 score:",r2)

```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR

dt=DecisionTreeRegressor(random_state=42)
dt.fit(x_train,y_train)
pre1=dt.predict(x_test)
dtr2=r2_score(y_test,pre1)
print("dt r2 score:",dtr2)

rf=RandomForestRegressor(n_estimators=100,random_state=42)
rf.fit(x_train,y_train)
pre2=rf.predict(x_test)
rfr2=r2_score(y_test,pre2)
print("rf r2 score:",rfr2)


sv=SVR(kernel="rbf")
sv.fit(x_train,y_train)
pre3=sv.predict(x_test)
svr2=r2_score(y_test,pre3)
print("sv r2 score:",svr2)

```
