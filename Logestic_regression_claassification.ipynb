```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("/content/hypertension_dataset.csv")
df.head()
```

```python
df.info()
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
lb=LabelEncoder()
ob=["BP_History","Medication","Family_History","Exercise_Level","Smoking_Status","Has_Hypertension"]
for col in ob:
  df[col]=lb.fit_transform(df[col])
df.head(10)
```

```python
df.info()
```

```python
df.shape
```

```python
x=df.drop("Has_Hypertension",axis=1)
y=df["Has_Hypertension"]

scaler = StandardScaler()
x = scaler.fit_transform(x)

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)

pre1=lr.predict(x_test)

as1=accuracy_score(y_test,pre1)
print("lg r accuracy:",as1)

knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
pre2=knn.predict(x_test)
as2=accuracy_score(y_test,pre2)
print("knn accuracy:",as2)

svm=SVC()
svm.fit(x_train,y_train)
pre3=svm.predict(x_test)
as3=accuracy_score(y_test,pre3)
print("svm accuracy:",as3)

nb=GaussianNB()
nb.fit(x_train,y_train)

pre4=nb.predict(x_test)

as4=accuracy_score(y_test,pre4)
print("nb accuracy:",as4)
```
